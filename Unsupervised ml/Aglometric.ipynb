```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
dataset=pd.read_csv(r"C:\Users\HP\OneDrive\Desktop\Machine learning\Iris.csv")
dataset.head(3)
```

```python
import scipy.cluster.hierarchy as sch
import matplotlib.pyplot as plt

# Drop the non-numeric column
x = dataset.drop(columns=['Species'])

# Plot the dendrogram
plt.figure(figsize=(30, 20))
plt.title("Dendrogram using Single Linkage")
plt.xlabel("Samples")
plt.ylabel("Euclidean Distance")

dendrogram = sch.dendrogram(
    sch.linkage(x, method="single", metric="euclidean")
)
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
ag=AgglomerativeClustering(n_clusters=2,linkage='single')
dataset["Predict"]=ag.fit_predict(x)
```
