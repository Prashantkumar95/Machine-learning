```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv("polynomiallogic.csv")
dataset.head(5)
```

```python
sns.scatterplot(x="x1",y="x2",data=dataset,hue="label")
```

```python
x=dataset.iloc[:,:-1]
y=dataset["label"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)*100
```

```python
from mlxtend.plotting import plot_decision_regions
```

```python
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=lr)
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python


pf = PolynomialFeatures(degree=2, include_bias=False)  # exclude bias if you don't want a column of 1s
X_poly = pf.fit_transform(x)

# Get correct column names
columns = pf.get_feature_names_out(x.columns)
x = pd.DataFrame(X_poly, columns=columns)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)*100
```
