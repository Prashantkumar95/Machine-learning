```python
import pandas as pd
```

```python
dataset=pd.read_csv("Social_Network_Ads.csv")
dataset.head(5)
```

```python
dataset["Purchased"].value_counts()
```

```python
x=dataset.iloc[:,:-1]
y=dataset["Purchased"]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
print(x_train.head())
print(x_train.dtypes)
```

```python
print(y_train.head())
print(y_train.dtypes)
```

```python
print(x_train.shape, y_train.shape)


```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
x_train['Gender'] = le.fit_transform(x_train['Gender'])
# 'Female' -> 0, 'Male' -> 1
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(x_train, y_train)
```

```python
x_train = pd.get_dummies(x_train, columns=['Gender'], drop_first=True)
```

```python
x_test = pd.get_dummies(x_test, columns=['Gender'], drop_first=True)

# Align columns to match training set
x_test = x_test.reindex(columns=x_train.columns, fill_value=0)
```

```python
y_test = y_test.map({'No':0, 'Yes':1})

```

```python
print(x_train.columns)
print(x_test.columns)

```

```python


# Make sure test set has the same columns as train set
x_test = x_test.reindex(columns=x_train.columns, fill_value=0)
```

```python
y_test = y_test.map({'No': 0, 'Yes': 1})
```

```python
print(x_train.shape, x_test.shape)
print(y_train.shape, y_test.shape)
```

```python
x_test = x_test.reindex(columns=x_train.columns, fill_value=0)

```

```python
print(x_train.columns)
print(x_test.columns)
```

```python
accuracy = lr.score(x_test, y_test) * 100
print(f"Test Accuracy: {accuracy:.2f}%")


```
